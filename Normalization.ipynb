# DataSet Creation


```python
df = pd.DataFrame({'Age':[20,30,40],
                     
                   'Income':[100000, 20000, 500000]})
  
display(df)
```

# Feature Scaling

```python
df['Income']= df['Income'] / df['Income'].max()
df
```

# Min - Max Method


```python
df['Income'] = (df['Income'] - df['Income'].min() ) / ( df['Income'].max() - df['Income'].min() )
df
```

# Z-Score (-3.0 ... 3.0)


```python
df['Income']= (df['Income'] - df['Income'].mean()) / (df['Income'].std())
df
```
